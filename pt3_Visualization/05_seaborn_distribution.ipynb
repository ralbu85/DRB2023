## Distribution Plot

```python
import seaborn as sns
sns.set_theme(style="darkgrid")
df = sns.load_dataset("penguins")
df
```

```python
sns.displot(df, x='flipper_length_mm')
```

```python
sns.displot(df, x='flipper_length_mm', bins=15)
```

## 범주에 따라 분포 분리하기

```python
sns.displot(df, x='flipper_length_mm', hue='species')
```

```python
sns.displot(df, x='flipper_length_mm', hue='species', element='step')
```

```python
# 구분해서 그리기
sns.displot(df, x='flipper_length_mm', col='species')
```

## 밀도함수로 정규화하여 집단간 분포비교 수행

```python
# 횟수가 중요한게 아니라 분포간의 동등한 비교를 하고싶을때
sns.displot(df, x='flipper_length_mm', hue='species', stat='density', common_norm=False)
```

```python
sns.displot(df, x="flipper_length_mm", kind="kde", hue='species', common_norm=False)
```

```python
sns.displot(df, x="flipper_length_mm", kind="kde", row='species', common_norm=False, height=2, aspect=4)
```

## 2차원 분포 표현

```python
sns.displot(df, x='bill_length_mm', y='bill_depth_mm')
```

```python
sns.displot(df, x='bill_length_mm', y='bill_depth_mm',col='species')
```

```python
sns.displot(df, x='bill_length_mm', y='bill_depth_mm',hue='species',kind='kde')
```
