## 피벗 테이블 만들기

* 다양한 축을 기준으로 데이터를 재구성하고, 집계함수를 적용하여 원하는 형태로 정리
    - 축: 범주형 데이터
    - 집계함수를 통해 수치형값을 요약

```python
import pandas as pd
df = pd.read_csv('data/credit.csv')
df
```

```python
df.pivot_table(index='Married', columns='Ethnicity', values='Cards', aggfunc='mean')
```

```python
df.pivot_table(index='Married',columns='Student',values='Income',aggfunc='mean')
```
