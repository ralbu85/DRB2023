## 데이터 합치기: Toy Example

```python
import pandas as pd

# Create sample DataFrame 1
data1 = {'ID': [1, 2, 3, 4, 5],
         'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve']}
df1 = pd.DataFrame(data1)

# Create sample DataFrame 2
data2 = {'ID': [1, 2, 3, 6, 7],
         'Age': [25, 32, 28, 40, 36]}
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

## 합치기

```python
pd.merge(df1, df2, on='ID')
```

* how = 'inner','left', 'right','outer'

```python
pd.merge(df1, df2, on='ID', how='left')
```

```python
pd.merge(df1, df2, on='ID', how='right')
```

```python
pd.merge(df1, df2, on='ID', how='outer')
```

## 실전예제

```python
import pandas as pd

airport = pd.read_csv('merge/L_AIRPORT_ID.csv')
carrier = pd.read_csv('merge/L_UNIQUE_CARRIERS.csv')
weekday = pd.read_csv('merge/L_WEEKDAYS.csv')
ontime = pd.read_csv('merge/T_ONTIME_MARKETING.csv')
```

```python
df2=pd.merge(ontime.fillna(0), carrier, left_on='OP_UNIQUE_CARRIER', right_on='Code',how='inner')
df2.groupby('Description').agg({'ARR_DEL15':'sum', 'DEP_DEL15':'sum'})
```
