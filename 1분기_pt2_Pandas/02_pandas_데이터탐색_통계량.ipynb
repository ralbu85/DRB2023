```python
import pandas as pd
df = pd.read_csv('data/credit.csv')
df
```

## 전체에 대한 요약값

```python
df.mean()
```

## 개별 칼럼에 대한 요약값

```python
print(df['Income'].mean())
print(df['Income'].max())
print(df['Income'].median())
print(df['Income'].min())
print(df['Income'].std())
```

## 수치형 데이터에 대하여 한눈에 요약통계량 보기
