## CNN 모델 만들기

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical

# MNIST 데이터셋을 로드하고 전처리합니다.
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)

# CNN 모델을 구축합니다.
model = Sequential()

# 합성곱 계층들을 추가합니다.
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# 특성 맵을 평탄화합니다.
model.add(Flatten())

# 완전 연결 계층들을 추가합니다.
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))  # 규제를 위한 드롭아웃
model.add(Dense(10, activation='softmax'))  # 10개 클래스를 가진 출력 계층

# 모델을 컴파일합니다.
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 모델을 훈련합니다.
history = model.fit(x_train, y_train, epochs=10, batch_size=128, validation_data=(x_test, y_test))

# 테스트 데이터에서 모델을 평가합니다.
score = model.evaluate(x_test, y_test, verbose=1)
print(f'테스트 손실: {score[0]}, 테스트 정확도: {score[1]}')
```

## 러닝커브 그리기

```python
import matplotlib.pyplot as plt

# Plotting the training and validation loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss Value')
plt.legend()
```

```python
# Plotting the training and validation accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy Value')
plt.legend()
```

## 오분류된 데이터 확인해보기


* y_pred = model.predict(x_test)은 모델이 테스트 데이터 x_test에 대한 예측을 수행한 결과를 저장합니다. 각 행은 입력 샘플에 대한 모델의 클래스별 예측 확률을 포함하고 있습니다.

* np.argmax(y_pred, axis=1)은 각 행에서 최대값(가장 높은 예측 확률)의 인덱스를 찾아내어 각 입력 샘플에 대한 예측된 클래스 레이블을 구합니다. 이것은 모델의 각 입력 샘플에 대한 예측된 클래스 레이블을 나타내는 배열을 생성합니다.



```python
# 예측 데이터 
y_pred = model.predict(x_test)
y_pred
```

```python
import numpy as np
y_pred_classes = np.argmax(y_pred,axis=1)
y_pred_classes
```

## 정답과 비교

```python
y_label = np.argmax(y_test,axis=1)
```

```python
correct_indices = np.where(y_pred_classes == y_label)[0]  # 올바르게 분류된 샘플의 인덱스
incorrect_indices = np.where(y_pred_classes != y_label)[0]  # 잘못 분류된 샘플의 인덱스
```

```python
incorrect_indices
```

## 오분류된 샘플 그려보기

```python
import numpy as np
import matplotlib.pyplot as plt

# incorrect_indices 중에서 무작위로 하나의 인덱스를 선택합니다.
idx = np.random.choice(incorrect_indices)

# 선택된 인덱스에 해당하는 이미지를 가져와서 시각화합니다.
plt.imshow(x_test[idx])
print(f"실제 레이블: {y_label[idx]}\n예측된 레이블: {y_pred_classes[idx]}")

```
