```python
import pandas as pd

df = pd.read_csv('data/bike.csv')
df
```

## 여러개의 차트를 동시에 만들기

Customization options include methods like 
* axes.set_title()
* axes.set_xlabel(), axes.set_ylabel()
* axes.set_xticks(), axes.set_yticks()
* axes.legend(), etc.

```python
import matplotlib.pyplot as plt

# 서브플롯 만들기
fig, axes = plt.subplots(2,1, figsize=(12,6), sharex=True)

# 첫번째 그래프
axes[0].plot(df['temp'])
axes[0].set_title('temperature')
axes[0].set_ylabel("temperature")
axes[0].grid()

# 두번째 그래프
axes[1].plot(df['hum'])
axes[1].set_title("humidity")
axes[1].set_ylabel("humidity")
axes[1].grid()

# 보이기
```

```python
# 서브플롯 만들기
fig, axes = plt.subplots(2,2, figsize=(12,8), sharex=True)

# 1행 1열
axes[0,0].hist(df['temp'])
axes[0,0].set_xlabel("humidity")
axes[0,0].set_ylabel("temperature")
axes[0,0].grid()

# 1행 2열
axes[0,1].scatter(df['hum'],df['temp'],s=1)
axes[0,1].set_xlabel("humidity")
axes[0,1].set_ylabel('temp')

# 두번째 그래프
axes[1,1].hist(df['hum'])
axes[1,1].set_ylabel("humidity")
axes[1,1].set_xlabel("temperature")
axes[1,1].grid()

# 2행 1열
axes[1,0].scatter(df['temp'],df['hum'],s=1)
axes[1,0].set_xlabel('temperature')
axes[1,0].set_ylabel('humidity')
```
