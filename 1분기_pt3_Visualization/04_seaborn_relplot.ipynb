## relplot

데이터 사이의 두 변수간의 관계를 그리는데 초점

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
tips
```

## relplot 예시

```python
sns.relplot(data=tips, x='total_bill', y='tip')
```

## seaborn의 핵심기능: 추가적 변수에 따른 구분

```python
# hue -> 카테고리별로 구분하여 데이터 표현 가능
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker')
```

```python
## 스타일까지 변경 가능
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker', style='smoker')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker', style='time')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='size')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='size', size='size')
```

## 시계열 데이터 표현

```python
bike = pd.read_csv('data/bike.csv')
bike
```

```python
bike.dtypes
```

```python
# 데이터의 형태를 명확히 시계열로 바꾸어줘야 잘 작동한다
bike['dteday'] = pd.to_datetime(bike['dteday'], format='%Y-%m-%d')
```

```python
# kind = 'line'
# 높이 너비 조절에 유의
sns.relplot(data=bike, x='dteday', y='temp', kind='line', height=6, aspect=1.5)
```
