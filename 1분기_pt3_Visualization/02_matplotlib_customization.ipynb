```python
import pandas as pd
df = pd.read_csv('data/bike.csv')
df
```

```python
import matplotlib.pyplot as plt
plt.plot(df['temp'])
```

## Matplotlib 조정하기

```python
import numpy as np

# 그림크기 조정하기
plt.figure(figsize=(20,8))

# 라인차트 그리기
plt.plot(df['temp'], label ='temp', color='green')
plt.plot(df['atemp'], label ='atemp', color='blue')

# 각 축에 제목달기
plt.xlabel('Days')
plt.ylabel('Temperature')

# 그래프 제목 달기
plt.title("Days vs Temperature")

# 틱값 조정하기
plt.xticks(np.arange(0,17500,5000))
plt.yticks(np.arange(0,1,0.1))

# 그리드 넣기
plt.grid()

# 축 한계치 조정하기
plt.xlim(0,18000)
plt.ylim(0,1)

# 범례 쓰기
plt.legend()

# 그리기
plt.show()

```
