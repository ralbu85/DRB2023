```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('data/auto.csv')
df
```

```python
# 데이터 형태를 명확히 설정해주는것이 좋다
df['origin'] = df['origin'].astype('category')
```

```python
sns.catplot(data=df, x='origin', y='weight')
```

```python
sns.catplot(data=df, x='origin', y='weight', jitter=False)
```

```python
sns.catplot(data=df, x='origin', y='weight', kind='swarm')
```

```python
sns.catplot(data=df, y='origin', x='weight', kind='swarm')
```

## boxplot으로 비교

```python
sns.catplot(data=df, x='origin', y='weight', kind='box')
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='box', hue='sex')
```

```python
sns.catplot(data=tips, x='total_bill', y='day', kind='violin', hue='sex')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', hue='sex', split=True)
```

## Point plots

```python
titanic = sns.load_dataset('titanic')
sns.catplot(data=titanic, x='sex', y='survived', hue='class', kind='point')
```
