## 아웃라이어 제거

```python
import pandas as pd
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
df=pd.read_csv(url,delimiter=';' )
```

## 아웃라이어 제거방법

1. 수치형 값에 대하여 표준정규분포로 표준화
    - z = (x - μ) / σ
    - z값의 절대값이 3이 넘어가는 데이터는 발생확률이 0.01% 미만으로 극단적인 데이터로 간주
2. z의 절대값이 3이 넘어가는 데이터의 행만 추리기 (any 사용)
3. 위 결과를 이용해 필터링

## 예시

```python
# 표준 정규분포 계산
z = (df - df.mean())/df.std()
```

```python
# z가 3이 넘는 경우 필터링
(z.abs() > 3)
```

```python
df[z.abs()<3]
```

```python
# any를 이용해 outlier가 하나라도 발생했던 행은 모두 제거
df[z.abs()<3].dropna()
```
