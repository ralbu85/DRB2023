## 인덱싱

```python
import pandas as pd
df = pd.read_csv('data/credit.csv')
df
```

## 인덱스를 리셋하기

현재 인덱스를 0부터 다시 주고 싶을때, 그리고 기존의 인덱스는 열로 보내고 싶을때

```python
df2 = df[df['Cards']>3]
df2
```

```python
df3 = df2.reset_index(names='old_idx')
```
