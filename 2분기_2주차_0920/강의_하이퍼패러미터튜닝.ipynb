```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ralbu85/DRB2023/master/Dataset/auto.csv')
df
```

```python
df = pd.get_dummies(df,columns=['origin'])
```

```python
y = df['mpg']
X = df.drop(columns=['mpg'])
```

## 하이퍼패러미터 차수로 조절

```python
from sklearn.pipeline import Pipeline # 데이터 변환 -> 선형회귀 학습
from sklearn.linear_model import LinearRegression, Ridge # 람다(패널티)를 다룰수 있는 선형회귀모형
from sklearn.preprocessing import PolynomialFeatures # 데이터를 고차원으로 변환
from sklearn.model_selection import KFold, GridSearchCV, cross_validate # 하이퍼패러미터 튜닝
```

```python
poly = PolynomialFeatures()
# model = LinearRegression()
model = Ridge()
```

```python
pipe = Pipeline([('poly',poly),('model',model)])
```

```python
degrees = range(1,8) # 차수 하이퍼패러미터 범위설정
la = [0.01,0.1,0.5,1,5,10] #alpha 하이퍼패러미터 범위 설정
param_grid = {'poly__degree':degrees, 'model__alpha':la}
```

```python
kfold = KFold(n_splits = 5, shuffle=True)
```

```python
grid_search = GridSearchCV(
    estimator = pipe,# 하이퍼패러미터 튜닝할 모델
    param_grid = param_grid,# 모델내의 하이퍼패러미터의 범주
    cv = kfold,#교차검증 방식
    scoring='r2',#성능지표
    return_train_score = True, #훈련데이터의 성능 반환 여부
)
```

```python
grid_search.fit(X,y)
```

```python
grid_search.best_params_
```
