## 6-7. continue

```python
for i in range(10):
    if i%2 == 0:
        continue
    print(i)
```

```python
for val in "Data":
    if val == "a":
        continue
    print(val)

print("The end")
```

```python
var = 5
while var > 0:
    var = var - 1
    if var == 2:
        continue
    print('Current variable value :', var)
print("Good bye!")
```

## 6-8. break

```python
for i in range(10):    # 0부터 9999까지 반복
    print(i)
    if i == 3:    # i가 100일 때
        break       # 반복문을 끝냄. for의 제어흐름을 벗어남
```

```python
i = 0
while True:    # 무한 루프
    print(i)
    i += 1          # i를 1씩 증가시킴
    if i == 12:    # i가 100일 때
        break       # 반복문을 끝냄. while의 제어흐름을 벗어남
```

```python
answer = 3
while True:
    x = int(input("입력해보세요:"))
    if(x==answer):
        print("맞았습니다.")
        break
    print("틀렸습니다.")
```
